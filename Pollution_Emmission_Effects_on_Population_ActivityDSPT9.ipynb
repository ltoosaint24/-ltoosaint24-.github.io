The Confidence Interval can be determine with analysis of standard error, and t-intervals. The intervals are statistical measures that valiate how sure we should be about the current data outcomes. The standard error, provides a value that measure possible discrepancies that can be found in data.

```
#Confidence Interval of Nitrogen Oxides
#Sample of Nitrogen Oxides
nitrogen_sample = air_pollution_gen['Nitrogen oxides (NOx)'].sample(50)
nitrogen_sample
```

```
#Mean of sample of Nitrogen Oxides
nitro_sample_mean = nitrogen_sample.mean()
nitro_sample_mean
```

The sample mean of Nitrogen Oxides was within the same proximity of the overall data mean, 9.7. This shows a validation that data occurrence within a consistent deviation.

```
#Standard deviation:
nitro_sample_std = nitrogen_sample.std()
nitro_sample_std
```

```
#The standard error is calculated to analysis for any statistical indication of data interuptions/discrepancy
#The count will be based on the sample size 50.
se_nitro_sample = nitro_sample_std/(len(nitrogen_sample)**1/2)
se_nitro_sample
```

```
#T-distribution
from scipy.stats import t
t_nitro = t.ppf(0.975, df = 49)
t_nitro
```

```
#The T-distribution use to calculate the lower and upper confidence limit, this means the range value that we
# can be confidene to expect to view within our data set about Nitrogen Dioxide
Lower_CL = nitro_sample_mean -(t_nitro * se_nitro_sample)
Lower_CL
#The lower confidence limit for Nitrogen Dioxide.
```

```
Upper_Cl = nitro_sample_mean + (t_nitro * se_nitro_sample)
Upper_Cl
#The lower confidence limit
```

```
#The t-interval provides the statistical validation regarding how confidence we should about the Nitrogen Dioxides values
#to always fall within these values.

t.interval(alpha = 0.95, df = 49, loc = nitro_sample_mean, scale = se_nitro_sample ) 
#The alpha value explains that we can be 95% confidence that Nitrogen Dioxide emissions will fall between 8.3943 and 10.006.
```
